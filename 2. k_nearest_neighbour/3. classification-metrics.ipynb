```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction-uci/heart.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
# Trainig above both model on same data

clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
# Prediction of both model using X_test

y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

### Accuracy : The percentage of instances that are correctly classified.

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print("Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees:",accuracy_score(y_test,y_pred2))
```

### Confusion Matrix

```python
confusion_matrix(y_test,y_pred1)
```

#### confusion matrix for multi-classification problem

```python
df1 = pd.read_csv('/kaggle/input/iris/Iris.csv')
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df1['Species'] = encoder.fit_transform(df1['Species'])
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1 = train_test_split(df1.iloc[:,0:-1],df1.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf3 = LogisticRegression()
clf4 = DecisionTreeClassifier()
```

```python
clf3.fit(X_train1,y_train1)
clf4.fit(X_train1,y_train1)
```

```python
y_pred3 = clf3.predict(X_test1)
y_pred4 = clf4.predict(X_test1)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression:",accuracy_score(y_test1,y_pred3))
print("Accuracy of Decision Trees:",accuracy_score(y_test1,y_pred4))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test1,y_pred3),columns=list(range(0,3)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test1,y_pred4),columns=list(range(0,3)))
```

### Recall, Precision, F1-Score

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print("For Logistic Regression Model")
print("-"*30)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,2)))
print(cdf)
print("-"*30)
print("Precision:",precision_score(y_test,y_pred1))
print("Recall:",recall_score(y_test,y_pred1))
print("F1 score:",f1_score(y_test,y_pred1))
```

```python
print("For Decision Tree Model")
print("-"*25)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,2)))
print(cdf)
print("-"*25)
print("Precision:",precision_score(y_test,y_pred2))
print("Recall:",recall_score(y_test,y_pred2))
print("F1 score:",f1_score(y_test,y_pred2))
```

##### In above score calculation, we are calculating for Positive(1) class, but in reality those functions are calculating for both the classes

```python
precision_score(y_test,y_pred1,average=None)
```

```python
precision_score(y_test,y_pred2,average=None)
```

```python
recall_score(y_test,y_pred1,average=None)
```

```python
recall_score(y_test,y_pred2,average=None)
```

```python
df2 = pd.read_csv('/kaggle/input/mnist-digit-recognizer/train.csv')
df2.head()
```

```python
df2.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.iloc[:,1:],df2.iloc[:,0],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf5 = LogisticRegression()
clf6 = DecisionTreeClassifier()
```

```python
clf5.fit(X_train,y_train)
clf6.fit(X_train,y_train)
```

```python
y_pred1 = clf5.predict(X_test)
y_pred2 = clf6.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression:",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees:",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```
