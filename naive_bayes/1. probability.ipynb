```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

### Joint Probability
- Use " normalize='all' " in Contingency Table code

##### Contingency Table
- X: Random Variable Associated with Pclass of Passenger.
- Y: Random Variable Associated with Survival Status of Passenger.

```python
pd.crosstab(df['Survived'], df['Pclass'])
```

```python
80/891
```

```python
pd.crosstab(df['Survived'], df['Pclass'],normalize='all')
```

### Marginal Probability
- Use " margins=True " in Contingency Table with margin code

##### Contingency Table with Margin
- X: Random Variable Associated with Pclass of Passenger.
- Y: Random Variable Associated with Survival Status of Passenger.

```python
pd.crosstab(df['Survived'], df['Pclass'],margins=True)
```

```python
pd.crosstab(df['Survived'], df['Pclass'],normalize='all',margins=True)
```

### Conditional Probability
- Use " normalize='columns' " in Contingency Table code

##### Probabilty of Dying/Alive when travelling in different class.

```python
pd.crosstab(df['Survived'], df['Pclass'],normalize='columns')
```

##### Probabilty of travelling in classes, when they have died/alive.

```python
pd.crosstab(df['Survived'], df['Pclass'],normalize='index')
```
