```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

### Applying Bagging Classifier

```python
# Training 10 DT base models

bag = BaggingClassifier(max_features=2,n_estimators=10)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
# printing 1 DT out of 10

plot_tree(bag.estimators_[0],feature_names=df.columns[:-1])
plt.show()
```

### Applying Random Forest Classifier

```python
rf = RandomForestClassifier(max_features=2,n_estimators=100)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
# printing 1 DT out of 10

plot_tree(rf.estimators_[4],feature_names=df.columns[:-1])
plt.show()
```
