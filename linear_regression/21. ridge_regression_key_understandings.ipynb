### 1. How are coefficients affected? -- Understand using graph

- Coefficient is shrinking towards zero, but not becoming zero when we increase alpha/lambda. 
- See scale of y-axis in graph below, as it is shrinking when aplha/lambda is increasing.

```python
from sklearn.datasets import load_diabetes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = load_diabetes()  
```

```python
# making dataframe
df = pd.DataFrame(data.data,columns=data.feature_names)
df['TARGET'] = data.target
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2,random_state=2)
```

##### Applying Ridge Regression on above data by gradually increasing lambda/alpha value

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
coefs = []
r2_scores = []

for i in [0,10,100,1000]:
    # creating object of Ridge Class
    reg = Ridge(alpha=i)
    # training model
    reg.fit(X_train,y_train)
    # storing value of coefficients in a list in coefs
    coefs.append(reg.coef_.tolist())
    y_pred = reg.predict(X_test)
    # storing value of R2-Scores in a list in r2_scores
    r2_scores.append(r2_score(y_test,y_pred))
```

```python
# Plot graph of all coefficients correspond to every alpha

plt.figure(figsize=(14,9))
plt.subplot(221)
plt.bar(data.feature_names,coefs[0])
plt.ylabel('alpha/lambda')
plt.title('Alpha = 0 ,r2_score = {}'.format(round(r2_scores[0],2)))

plt.subplot(222)
plt.bar(data.feature_names,coefs[1])
plt.title('Alpha = 10 ,r2_score = {}'.format(round(r2_scores[1],2)))

plt.subplot(223)
plt.bar(data.feature_names,coefs[2])
plt.ylabel('alpha/lambda')
plt.title('Alpha = 100 ,r2_score = {}'.format(round(r2_scores[2],2)))

plt.subplot(224)
plt.bar(data.feature_names,coefs[3])
plt.title('Alpha = 1000 ,r2_score = {}'.format(round(r2_scores[3],2)))

plt.show()
```

### 3. Impact on Bias and Variance

```python
# Generating Non-Linear Data

m = 100
X = 5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

def get_preds_ridge(X, y, alpha):
    model = Pipeline([
        ('poly_feats', PolynomialFeatures(degree=16)),
        ('ridge', Ridge(alpha=alpha))
    ])
    model.fit(X, y)
    return model.predict(X)

alphas = [0, 20, 200]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10, 6))
plt.plot(X, y, 'b+', label='Datapoints')

for alpha, c in zip(alphas, cs):
    preds = get_preds_ridge(X, y, alpha)
    # Plot
    plt.plot(sorted(X[:, 0]), preds[np.argsort(X[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.reshape(100,1),y.reshape(100),test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=15)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
# Using bias_variance_decomp function of mlxtend library to calculate Bias & Variance
# bias_variance_decomp function -- Gives Bias & Variance value for a particular alpha/lambda value.

from mlxtend.evaluate import bias_variance_decomp
# taking 100 values of alpha b/w 0 to 40, & for every value calculate Bias & Variance & store in list.
alphas = np.linspace(0,25,100)
loss = []
bias = []
variance = []

for i in alphas:
    reg = Ridge(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(reg, X_train, y_train, X_test, y_test, loss='mse',random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
# Plotting Graph to see what will happen to Bias & Variance when we increase alpha/lambda

plt.plot(alphas,loss,label='loss')
plt.plot(alphas,bias,label='Bias')
plt.plot(alphas,variance,label='Variance')
plt.ylim(0,5)
plt.xlabel('Alpha')
plt.legend()
plt.show()
```
